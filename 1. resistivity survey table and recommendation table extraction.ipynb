```python
def load_pkl(file):
    with open(file, 'rb') as f: 
        var = pickle.load(f)
        f.close()
    return var
def save_pkl(file,var):
    with open(file, 'wb') as f:  # Python 3: open(..., 'wb')
        pickle.dump(var, f)
        f.close()

#             print(a[0].strip(),a[1].replace('-','').strip() )
#         if(len(a)==3):
#             fields.append(a[0].strip())
#             vals.append(a[1].replace('-','').replace('Distt','').strip())
#             fields.append('Distt')
#             vals.append(a[2].replace('-','').strip())
#             print(a[0].strip(),a[1].replace('-','').strip() )

    
def get_cell_data(cells):
    celldata=[]
    for cell in cells:
        for paragraph in cell.paragraphs:
            if 'Distt' in paragraph.text:
                splittext=paragraph.text.split('Distt')
                splittext=[splittext[0] ,'Distt'+splittext[1]]
            else:
#                 splittext=paragraph.text.split('   ')
                splittext=paragraph.text.split('   ')

#             print(splittext)
            celldata=np.append(celldata,[c.strip() for c in splittext])        
    return celldata
def cols_seeNadd(cols1,cols2):
#     cols=cols1.copy()
    for i in range(len(cols1)):
        if(cols1[i]==cols2[i]):
            cols1[i]=np.core.defchararray.add(cols1[i],cols2[i])
    return cols1
            
def get_data_frm_interpTable(table,datatype='all',headcount=2):
    if(datatype=='all' or datatype=='hdata'):
        if(headcount==2):
            hcols=cols_seeNadd(get_cell_data(table.rows[0].cells), get_cell_data(table.rows[1].cells))
        else:
            hcols=get_cell_data(table.rows[0].cells)
        if (datatype=='hdata'):
            return hcols
    if(datatype=='all' or datatype=='ddata'):
        dcols=[]
        for row in table.rows[headcount:-1]:
#             dcols=np.append(dcols,get_cell_data(row.cells))
            dcols.append(get_cell_data(row.cells))
        if(datatype=='ddata'):
            return dcols
    if(datatype=='all' or datatype=='rdata'):
        last_col=np.unique(get_cell_data(table.rows[-1].cells))
        if(datatype=='rdata'):
            return last_col
    return hcols,dcols,last_col  


# #         print()
# table.rows[-1].cells[0].paragraphs[0].text
# #header of table part
# hcols=np.ndarray(shape=(1,len(table.rows[0].cells)), dtype='U100').ravel()
# dcols=[]
# # cols = np.dtype(('U10',(1,len(table.rows[0].cells))))
# for row in table.rows[0:2]:
#     hcols=cols_seeNadd(hcols, get_cell_data(row.cells))
# #     print(np.core.defchararray.add(cols, get_cell_data(row.cells)))
# for row in table.rows[2:-1]:
#     dcols=np.append(dcols,get_cell_data(row.cells))
#     print()
# last_col=np.unique(get_cell_data(table.rows[-1].cells))

# hcols,dcols,last_col  
```
